```python
import matplotlib.pyplot as plt
import re
```

```python
# Открываем файл
filename = "Уэллс_-_Война_миров.txt"
with open(filename, "r") as file:
    text = file.read()
```

```python
# Подсчет слов, строк и символов с пробелами
word_count = len(re.findall(r'\w+', text))
line_count = text.count("\n") + 1
character_count = len(text)
# Подсчет символов без пробелов
text_without_spaces = text.replace(" ", "")
character_count_without_spaces = len(text_without_spaces)
# Построение гистограммы наиболее употребительных слов
word_frequency = {}
words = re.findall(r'\w+', text.lower())
for word in words:
    if word in word_frequency:
        word_frequency[word] += 1
    else:
        word_frequency[word] = 1
sorted_words = sorted(word_frequency.items(), key=lambda x: x[1], reverse=True)
top_words = sorted_words[:20]
labels = [word[0] for word in top_words]
counts = [word[1] for word in top_words]
```

```python
plt.bar(labels, counts)
plt.xticks(rotation=90)
plt.xlabel("Слова")
plt.ylabel("Частота")
plt.title("Гистограмма наиболее употребительных слов")
plt.tight_layout()
plt.show()

# Вывод сводки
print("Сводка файла:")
print("Количество слов: ", word_count)
print("Количество строк: ", line_count)
print("Количество символов (с пробелами): ", character_count)
print("Количество символов (без пробелов): ", character_count_without_spaces)
```
